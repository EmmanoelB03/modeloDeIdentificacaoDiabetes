```python
import pandas as pd
import matplotlib.pyplot as plt
import kagglehub
```

```python
path = kagglehub.dataset_download('alexteboul/diabetes-health-indicators-dataset')

print(path+'/diabetes_binary_health_indicators_BRFSS2015.csv')
```

```python
df = pd.read_csv(path+'/diabetes_binary_health_indicators_BRFSS2015.csv')
```

```python
df.head()
```

```python
#Analize dataSet

#Diabetes_binary: Has diabetes; 1 -> Yes; 0 -> No;
#HighBP: Has high blood pressure; 1 -> Yes; 0 -> No;
#HighChol: Has high cholesterol; 1 -> Yes; 0 -> No;
#CholCheck: Had cholesterol check in the last 5 years; 1 -> Yes; 0 -> No;
#BMI: Body Mass Index; Underweight: Less than 18.5; Normal/Healthy weight: 18.5 to 24.9; Overweight: 25.0 to 29.9; Obesity Class I: 30.0 to 34.9; Obesity Class II: 35.0 to 39.9; Obesity Class III: 40.0 or higher;
#Smoker: Has smoked at least 100 cigarettes in lifetime; 1 -> Yes; 0 -> No;
#Stroke: Had a stroke; 1 -> Yes; 0 -> No;
#HeartDiseaseorAttack: Has coronary heart disease or heart attack; 1 -> Yes; 0 -> No;
#PhysActivity: Has physical activity in the last 30 days; 1 -> Yes; 0 -> No;
#Fruits: Consumes fruits 1 or more times per day; 1 -> Yes; 0 -> No;
#Veggies: Consumes vegetables 1 or more times per day; 1 -> Yes; 0 -> No;
#HvyAlcoholConsump: Heavy alcohol consumption (>=14 drinks per week for men, >=7 for women); 1 -> Yes; 0 -> No;
#AnyHealthcare: Has healthcare coverage; 1 -> Yes; 0 -> No;
#NoDocbcCost: Couldn't see a doctor due to cost; 1 -> Yes; 0 -> No;
#GenHlth: General health self-assessment; 1 -> Excellent; 2 -> Very Good; 3 -> Good; 4 -> Fair; 5 -> Poor;
#MentHlth: Days of poor mental health in the last 30 days; range -> 0 to 30;
#PhysHlth: Days of poor physical health in the last 30 days; range -> 0 to 30;
#DiffWalk: Has great difficulty walking/climbing stairs; 1 -> Yes; 0 -> No;
#Sex: Biological sex; 0 -> Female; 1 -> Male;
#Age: Age category; range -> 1 to 13;
#Education: Education level; range -> 1 to 6;
#Income: Annual household income level; range -> 1 to 8;
```

```python
contagem_por_idade = df['Age'].value_counts().sort_index()

# Criando o gráfico de barras
plt.figure(figsize=(10,5))

countElements = df['Age'].value_counts().sort_index()
#cada count que passsa ele verifica o index(o valor armazenado naquele count como se fosse um dicionario)
plt.bar(countElements.index.astype(str), countElements, color='skyblue')

plt.title('Age destribution')
plt.xlabel('age')
plt.ylabel('Count')
plt.show()

```

```python
#BMI/AGE

countBmiElements = df['BMI'].value_counts().sort_index()

plt.figure(figsize=(20,6))
plt.bar(countBmiElements.index.round(3).astype(str), countBmiElements, color='red')
plt.title('BMI destribution')
plt.xlabel('BMI')
plt.xticks(rotation=45,ha='right')
plt.ylabel('Counts')
plt.show()
```

```python
plt.figure(figsize=(18,6))
plt.scatter(df["BMI"].index,df["BMI"])
plt.xlabel('ID')
plt.ylabel('BMI')
plt.title('BMI Destribution')
plt.grid(axis='y', linestyle='--', alpha=0.7, color='black')
plt.show()
```

```python
# Identify outliers

#Quantile method

Q1 = df['BMI'].quantile(0.25)
Q3 = df['BMI'].quantile(0.75)

IQR = Q3 - Q1

lowerBound = Q1 - 1.5*IQR
upperBound = Q3 + 1.5*IQR

outliers = df[(df['BMI'] < lowerBound) | (df['BMI'] > upperBound)].copy()

#remove outliers from dataframe
df = df[(df['BMI'] >= lowerBound) & (df['BMI'] <= upperBound)].copy()

print(f"\nQ1: {Q1:.2f}, Q3: {Q3:.2f}, IQR: {IQR:.2f}")
print(f"Lower Bounds: {lowerBound:.2f}, Upper Bounds: {upperBound:.2f}")
print(f"Outliers identificados pelo método do IQR:")
print(outliers['BMI'])
print(outliers['BMI'].value_counts)


plt.scatter(outliers['BMI'].index, outliers["BMI"])
```

```python
#treat outliers as mean

mean = df['BMI'].mean()

outliers['BMI'] = mean

outliersTreat = outliers

plt.scatter(outliers['BMI'].index, outliers['BMI'])
plt.xlabel('ID')
plt.ylabel('BMI')
plt.title('Outliers BMI Destribution ')
plt.show()
```

```python
#add treat outliers in dataframe

plt.figure(figsize=(15,6))

countElements = df['BMI'].value_counts().sort_index()

plt.bar(countElements.index.round(3).astype(str), countElements, color='red')
plt.xlabel('BMI')
plt.xticks(rotation=45, ha='right')
plt.ylabel('ID')
plt.title('BMI without treat outliers')
plt.show()

df = pd.concat([outliersTreat,df]) #concat trat outliers

countElements = df['BMI'].value_counts().sort_index()

plt.figure(figsize=(15,6))

plt.bar(countElements.index.round(3).astype(str), countElements, color='skyblue')
plt.xlabel('BMI')
plt.xticks(rotation=45, ha='right')
plt.ylabel('ID')
plt.title('BMI with treat outliers')
plt.show()

```
